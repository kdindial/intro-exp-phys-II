```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#For 3d plots
from mpl_toolkits.mplot3d import Axes3D
```

```python
x0 = 1.
y0 = 3.
deltax=0.1
deltay=0.2
x = np.random.normal(x0,deltax,100000)
y = np.random.normal(y0,deltay,1000)
```

```python
q = 1/x
```

```python
q0=x0+y0
deltaq = np.sqrt( (deltax)**2 +  (deltay)**2 )
q0, deltaq
```

```python
np.mean(q), np.std(q)
```

```python
np.max(q)
```

```python
import seaborn as sns
```

```python
g = sns.JointGrid(x,q)
g.set_axis_labels('x','q')
g = g.plot_joint(plt.scatter, alpha=0.5)
g = g.plot_marginals(sns.distplot)
```

```python
g = sns.JointGrid(q,x)
g.set_axis_labels('q','x')
g = g.plot_joint(plt.scatter, alpha=0.5)
g = g.plot_marginals(sns.distplot)
```
